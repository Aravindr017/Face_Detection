# Object Detection

```python
# importing Libraries
import cv2
from google.colab.patches import cv2_imshow
```

```python
# using pretrained model

face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
```

```python
# reading image
image = cv2.imread('/content/drive/MyDrive/ICT - Ai Ml/Deep Learning/Convolutional Neural Network/Images/group_img.jpeg')
```

```python
# converting to grayscale

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
# detecting multiple faces in the image using pretrained model

faces = face_cascade.detectMultiScale(gray, 1.3, 1)
print(faces)    # return the detected edge array
```

```python
# for showing rectangle box outline for detected image

for (x, y, w, h) in faces:
  cv2.rectangle(image, (x,y), (x+w, y+h), (255,0,0), 2)
```

```python
# to display the face detected image
cv2_imshow(image)
```
